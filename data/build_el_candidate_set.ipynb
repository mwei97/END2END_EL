```python
import torch
import json
import numpy as np
```

```python
fpath = '../models/all_entities_large.t7'
```

```python
all_entities = torch.load(fpath)
```

```python
data_path = 'AIDA-YAGO2_longformer/tokenized/'

splits = ['train', 'dev', 'test']

# read samples
samples = []

for split in splits:
    with open(data_path+split+'.jsonl') as f:
        for line in f.readlines():
            samples.append(json.loads(line))
```

```python
len(samples)
```

```python
samples[0].keys()
```

```python
all_labels = []

for sample in samples:
    labels = sample['label_id']
    all_labels.extend(labels)
```

```python
len(all_labels)
```

```python
unique_labels = set(all_labels)
len(unique_labels)
```

```python
unique_labels = sorted(unique_labels)
```

```python
unique_labels
```

```python
all_entities.size()
```

```python
indices = torch.tensor(unique_labels)
indices.size()
```

```python
selected_entities = torch.index_select(all_entities, 0, indices)
```

```python
selected_entities.size()
```

```python
selected_entities
```

```python
torch.save(selected_entities, 'el_candidate_set.t7')
```

```python
idx2label = {i:lab for i,lab in enumerate(unique_labels)}
```

```python
idx2label
```

```python
with open('candidate_set_idx2label.json', 'w') as f:
    json.dump(idx2label, f)
```
